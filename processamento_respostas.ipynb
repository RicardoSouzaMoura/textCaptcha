```
df = pd.read_csv("respostas_pesquisa.csv", header=None, skiprows=1)
```

```
df = df.drop(df.columns[[0]], axis=1)
```

```
def temresposta(textoToSearch, textoFromSearch):
  if textoFromSearch.count(textoToSearch) > 0:
    return 1
  return 0
```

```
print(temresposta("abc", ["kmdfabckmdfa", "fsdf", "dfadf"] ))
print(temresposta("abc", ["sdfas", "abckmdfabdfsadckmdfa", "fafsd"]))
print(temresposta("abc", ["sdfsf", "abc", "abdfsckmdfabdfsadckmdfabc"]))
print(temresposta("abc", ["abdfsckmdfabdfsadckmdfabsdfsc"]))
print(temresposta("abc", ["'abc'"]))
```

```
df_result = pd.DataFrame(columns = ["usuario", "pergunta", "resposta", "tipo", "gabarito", "rotulo"])
for ind in df.index:
  for n in range(6):
    for k in range(5):
      data = {
        "usuario": ind,
        "pergunta": n,
        "resposta": dict.get('respostas')[n][k],
        "tipo": dict.get('tipo')[n][k],
        "gabarito": dict.get('gabarito')[n][k],
        "rotulo": temresposta(dict.get('respostas')[n][k], df[n+1][ind])
      }
      df_temp = pd.DataFrame(data, columns = ["usuario", "pergunta", "resposta", "tipo", "gabarito", "rotulo"] , index=[0])
      df_result = df_result.append(df_temp, ignore_index=True)
  break
```
